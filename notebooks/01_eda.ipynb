```python
import pandas as pd
import matplotlib as plt

df = pd.read_csv("../data/raw/co2.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(df["CO2 Emissions(g/km)"], bins=40, color='blue', alpha=0.7)
plt.xlabel("CO2(g/km)")
plt.ylabel("Count")
plt.title("Distribution of CO2 Emissions")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.scatter(df["Fuel Consumption Comb (L/100 km)"],
            df["CO2 Emissions(g/km)"], alpha=0.2)
plt.xlabel("Fuel Consumption Comb (L/100 km)")
plt.ylabel("CO2 Emissions (g/km)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["Engine Size(L)"], df["CO2 Emissions(g/km)"], alpha=0.2)
plt.xlabel("Engine Size (L)")
plt.ylabel("CO2 g/km")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["Cylinders"], df["CO2 Emissions(g/km)"], alpha=0.3)
plt.xlabel("Cylinders")
plt.ylabel("CO2 Emissions (g/km)")
plt.title("Cylinders vs CO2 Emissions")
plt.show()
```
